# now using MAjority voting technique to determine final prediction using ensemble learning

```python


def ensemble_row(list):
    test_data = pd.DataFrame(list, columns=['Outlook', 'Temperature', 'Humidity', 'Wind'])   # 1 - row and 4 - feature
    result1 = DC1.predict(test_data)
    result2 = DC2.predict(test_data)
    result3 = DC3.predict(test_data)
    predictions = [result1[0], result2[0], result3[0]]
    final = max(set(predictions), key=predictions.count)
    #print(final)
    if(final == 1):
        print('Player will play Tennis')
    else:
        print('Player will not play tennis')

list = [[2,2,2,2]]
ensemble_row(list)
```

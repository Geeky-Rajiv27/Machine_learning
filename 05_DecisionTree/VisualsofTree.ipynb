```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
df = load_iris()
```

```python
target = df.target
target
```

```python
features = df.feature_names
features
```

```python
X = df.data
y = df.target
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
DTC = DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split
X_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
DTC.fit(X_train, y_train)
```

```python
x_test
```

```python
y_pred = DTC.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(f"The accuracy of the decision tree based classfier is : {accuracy_score(y_test, y_pred):.5f}")
```

```python
#NOTE: NOw checking what was the tree drawn for this classificaiton 
from sklearn.tree import plot_tree
```

# NOTE: now making a prediction section where this model will predict based on input given by user

```python
def user_input():
    sepal_length = float(input("Enter sepal length in cm : "))
    sepal_width = float(input("enter sepal_width in cm : "))
    petal_length = float(input("Enter petal length in cm : "))
    petal_width = float(input("Enter petal width in cm : "))
    
    return sepal_length,sepal_width,petal_length,petal_width
```

```python
sepal_length,sepal_width,petal_length,petal_width = user_input()
```

```python
data = np.array([sepal_length,sepal_width,petal_length,petal_width])
```

```python
data = data.reshape(1,4)
```

```python
prediction = DTC.predict(data)
```

```python
if(prediction == 0):
    print("The species of flower is setosa")
elif(prediction == 1):
    print("The species of flower is versicolor")
else: 
    print("The species of flower is verginica")
```

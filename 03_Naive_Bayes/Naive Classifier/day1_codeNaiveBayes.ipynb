# 🟢 Scenario 1: Weather App Decision System

You are building a small ML system for a sports academy.
Every morning, the system predicts whether the academy should schedule outdoor tennis practice.

🌤 New Weather Conditions:

Outlook = Rain

Temperature = Cool

Humidity = Normal

Wind = Strong

❓ Question:

Using Naive Bayes, should the academy schedule tennis practice?

👉 Calculate:

P(Yes | conditions)

P(No | conditions)

Then decide the final class.

```python
py = prob_yes * prob_rain_yes * prob_Cool_yes * prob_Normal_yes * prob_Strong_yes 
print(f"The probability of yes is {py}")

pn = prob_yes * prob_rain_no * prob_Cool_no * prob_Normal_no * prob_Strong_no 
print(f"The probability of No is {pn}")
```

```python
if(py > pn):
    print("Players will play tennis !")
elif(py == pn):
    print("There is equal chance player will play tennis !")
else:
    print("Players won't play tennis !")
```

```python
# NOTE: Removing html tags using regex 
import re
clean = re.compile('<.*?>')
re.sub(clean, '', df.iloc[1].review)    # removed all the html tags of row no 2 (index 1)
```

```python
# NOTE : here i will remove now for all 50000 rows
def clean_tags(text):
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)
```

```python
# NOTE: callign above funciton
df['review'] = df['review'].apply(clean_tags)
df['review']
```

```python
# NOTE : creating special character remover function
def special_char(text):
    x = ''
    for word in text:
        if word.isalnum():
            x = x + word
        else:
            x = x + ' '     #put space if it is special characters
    return x
```

```python
#NOTE : calling above function initially for one test 
special_char('Raji@v# is a G*&^d Boy')
```

```python
# NOTE : Now i will call this for whole review rows 
df['review'] = df['review'].apply(special_char)
df['review']
```

```python
#NOTE : making lower case 
def lower_case(text):
    return text.lower()
```

```python
df['review'] = df['review'].apply(lower_case)
df
```

```python
#NOTE : Removing all stops words like is , are , has , have etc.
import nltk     #nltk : natural language toolkit
from nltk.corpus import stopwords
nltk.download('stopwords')          #requires for first time
```

```python
stopwords = stopwords.words('english')
```

```python
def Eng_stopwords_counts(words):
    count = 0
    for i in words: #Note ; words is list
        count = count + 1
    return count
    
```

```python
Eng_stopwords_counts(stopwords)
```

```python
print(type(df['review'][0]))
```

```python
#NOTE : there are total 198 stop words in 'english' language we gonna use for our sentiment analysis
def remove_stopwords(reviews):
    review_list = []
    for word in reviews.split():

        if word not in stopwords:
            review_list.append(word)    # if not a stop word in the review row list then only keep
        
        temp_review_list = review_list[:]
    review_list.clear()
    return temp_review_list
```

```python
#NOTE :calling above funciton : to remove stop words from all 50000 rows of review
df['review'] = df['review'].apply(remove_stopwords)
df
```

```python
#NOTE: replacing positive with 1 and negative with 0
df['sentiment'].replace({
    'positive' : 1,
    'negative' : 0
},inplace=True)
```

```python
df
```

```python
#NOTE :performing stemming using count_vectorizer()
from nltk.stem.porter import PorterStemmer
PS = PorterStemmer()    # calling the function
```

```python
row = []
def word_stemmer(reviews_row):
    for word in reviews_row:
        row.append(PS.stem(word))   # only storing stem words in each review list
    temp_list = row[:]
    row.clear()
    return temp_list    
```

```python
#NOTE: testing above function for single sentence
word_stemmer(['I', 'love', 'being', 'loved', 'by', 'everyone'])
```

```python
word_stemmer(['playing', 'eating', 'scoring'])
```

```python
#NOTE: Now running word stemmer funciton for all reviews
df['review'] = df['review'].apply(word_stemmer)
df
```

```python
#NOTE : now converting datatype of each review from list to string (sentence)
df['review'] = df['review'].apply(lambda row: " ".join(row))
```

# NOTE: text cleaning all completed

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=1000)         # words which are repeated atleast 1000 times
```

```python
X = cv.fit_transform(df['review']).toarray()
```

```python
X       #NOTE: this has 50000 rows and 1000 unique words in review
```

```python
X.mean()        #NOTE: most of the words are repeated so mean is almost equal to 0
```

```python
#NOTE : Y = (sentiments)
Y = df.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size= 0.2)
```

```python
from sklearn.naive_bayes import GaussianNB,BernoulliNB,MultinomialNB

Model_Gauss = GaussianNB()
Model_Bernauli = BernoulliNB()
Model_Multi = MultinomialNB()
```

```python
#NOTE: Training sets

Model_Gauss.fit(X_train, Y_train)
Model_Bernauli.fit(X_train, Y_train)
Model_Multi.fit(X_train, Y_train)
```

```python
#NOTE: Predicting sentiments [Testing Phase]
pred_sentiments_Gauss = Model_Gauss.predict(X_test)
pred_sentiments_Bernauli = Model_Bernauli.predict(X_test)
pred_sentiments_Multi = Model_Multi.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(f"The accuracy of Gaussian Model is {accuracy_score(Y_test,pred_sentiments_Gauss)}")
print(f"The accuracy of Bernauli model is : {accuracy_score(Y_test,pred_sentiments_Bernauli)}")
print(f"The accuracy of Multinomial model is {accuracy_score(Y_test,pred_sentiments_Multi)}")
```

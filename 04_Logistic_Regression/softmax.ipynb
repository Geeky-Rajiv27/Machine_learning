```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('penguins')
```

```python
df = df.drop(columns=['bill_depth_mm', 'body_mass_g', 'island', 'sex'])
```

```python
df.info()
```

```python
df.dropna()
```

```python
#NOTE: One hot label encoding
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
df['species'] = encoder.fit_transform(df['species'])
df['species']
```

```python
df.head()
```

```python
df = df.dropna()
df.head(10)
```

```python
X = df.iloc[:,1:]
Y = df.iloc[:, 0:1]
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression(multi_class='multinomial')  #NOTE : since softmax so multinomial
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=41)
```

```python
LR.fit(X_train,Y_train)
```

```python
pred_y = LR.predict(X_test)
```

```python
#NOTE: metrics
from sklearn.metrics import accuracy_score
```

```python
print(f"Accuracy of the model is : {accuracy_score(Y_test,pred_y)}")
```

```python
z = np.array([[3.4, 2.7]])      #NOTE: random points 
LR.predict_proba(z)     #NOTE : predicting probability using softmax funciton
```
